```python
import pandas as pd
```

```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}
```

```python
students_df= pd.DataFrame(students_data)
enrollments_df = pd.DataFrame(enrollments_data)
scores_df = pd.DataFrame(scores_data)
```

```python
print("=== TASK 1: Data Preparation ===")
print("Original Students DataFrame:")
print(students_df)
print("...")
print("Null Value Analysis:")
for col in students_df.columns:
    null_count = students_df[col].isnull().sum()
    null_percent = (null_count / len(students_df)) * 100
    print(f"Column: {col}, Nulls: {null_count} ({null_percent:.2f}%)")
```

```python
students_clean = students_df.copy()
students_clean['city'] = students_clean['city'].fillna('Unknown')
students_clean = students_df.dropna(subset=['name'])
print(f"Cleaned Students DataFrame ({len(students_clean)} rows):")
print(students_clean)
```

```python
print("=== TASK 2: Join Operations ===")
#Inner join students and enrollments
inner_join_df = pd.merge(students_df, enrollments_df, on='student_id', how='inner')
print(f"Inner Join Result ({len(inner_join_df)} rows):")
print(inner_join_df)
excluded_students = students_df.loc[
    ~students_df['student_id'].isin(enrollments_df['student_id']),
    'student_id'
].tolist()

print(f"\nExcluded students: {excluded_students} - Not in enrollments table")
```

```python
#left join students and enrollments
left_join_df = pd.merge(students_df, enrollments_df, on='student_id', how='left')
print(f"Left Join Result ({len(left_join_df)} rows):")
print(left_join_df)
students_with_null_course = left_join_df.loc[
    left_join_df['course_name'].isnull(), 'student_id'
].tolist()

print(f"\nStudents with null course_name: {students_with_null_course}")
```

```python
#right join students and enrollments
right_join_df = pd.merge(students_df, enrollments_df, on='student_id', how='right')
print(f"Right Join Result ({len(right_join_df)} rows):")
print(right_join_df)
enrollments_without_students = right_join_df.loc[
    right_join_df['name'].isnull(), 'student_id'
].tolist()

print(f"\nEnrollments with no matching student record: {enrollments_without_students}")
```

```python
#full outer join students and enrollments
full_outer_join_df = pd.merge(students_df, enrollments_df, on='student_id')
print(f"Full Outer Join Result ({len(full_outer_join_df)} rows):")
print(full_outer_join_df)
students_without_enrollments = full_outer_join_df.loc[
    full_outer_join_df['course_name'].isnull(), 'student_id'
].tolist()

# Enrollments without students
enrollments_without_students = full_outer_join_df.loc[
    full_outer_join_df['name'].isnull(), 'student_id'
].tolist()

print(f"\nStudents without enrollments: {students_without_enrollments}")
print(f"Enrollments without students: {enrollments_without_students}")
```

```python
merged_df = pd.merge(students_df, enrollments_df, on='student_id',how='outer',indicator=True)
print(f"Merged Join Result ({len(merged_df)} rows):")
print(merged_df)
print(merged_df['_merge'].value_counts())
```

```python
print("\n=== TASK 3: Lookup and Automation ===")
#Lookup using map
score_lookup = dict(zip(scores_df['student_id'], scores_df['exam_score']))

students_with_scores = students_df[['student_id', 'name']].copy()
students_with_scores['exam_score'] = students_with_scores['student_id'].map(score_lookup)

print("\nLookup Operation Result:")
print(students_with_scores)
```

```python
#same operation using merge
students_merge_scores = pd.merge(
    students_df[['student_id','name']],
    scores_df,
    on='student_id',
    how='left'
)

print("\nMerge Operation Result:")
print(students_merge_scores)
```

```python
'''map() advantages in this scenario:

Works like a dictionary lookup → O(1) complexity

Faster when adding a single column based on key

Less memory usage and Simpler syntax

merge() advantages:

Needed when joining multiple columns

Needed when merging multiple datasets

In this case we only need one value (exam_score) using one key (student_id), so .map() is faster.'''
```

```python
#Automation Function
def auto_merge(df1, df2, join_type, key_column):
    merged_df = pd.merge(df1, df2, on=key_column, how=join_type)

    return {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }
```

```python
#Automation function test

# 1. Inner Join

result_inner = auto_merge(students_df, enrollments_df, 'inner', 'student_id')

print("\nAutomation Function Test:")
print(f"Join Type: {result_inner['join_type']}")
print(f"Rows in Result: {result_inner['row_count']}")
print("Result Preview:")
print(result_inner['result_df'][['student_id','name','course_name']].head())
```

```python
# 2. Left Join

result_left = auto_merge(students_df, enrollments_df, 'left', 'student_id')

print(f"\nJoin Type: {result_left['join_type']}")
print(f"Rows in Result: {result_left['row_count']}")
print("Result Preview:")
print(result_left['result_df'][['student_id','name','course_name']].head())
```
